```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)

y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-90)
```

Tanh function

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)

y = tanh(x)

plt.plot(x, y, color='orange')
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.show()
```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(100)
```

```python
tanh(1)
```

```python
tanh(-90)
```

ReLU Function

-> Converts negative x to 0

-> Converts positive x to x itself

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)

y = relu(x)

# Plot the graph
plt.plot(x, y, color='green')
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.show()
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-7)
```

```python
relu(6)
```

Leaky ReLU Function

-> Convert negative x to 0.1 * x

-> Converts positive x to  x itself

```python
def leaky_relu(x):
    return np.where(x > 0, x, 0.1 * x)

x = np.linspace(-10, 10, 100)

y = leaky_relu(x)

plt.plot(x, y, color='purple')
plt.title("Leaky ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True)
plt.show()
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-7)
```

```python
leaky_relu(7)
```
