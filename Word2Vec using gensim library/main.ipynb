```python
import pandas as pd
import gensim
```

We are using a cell phones and accessories review dataset of Amazon.

We will apply the word2vec functionalities on the reviews of this dataset and create a model

```python
df = pd.read_json('Cell_Phones_and_Accessories_5.json', lines=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.reviewText[0]
```

Now we will create clean tokens of these review using simple_process function of gensim.
 
Its main job is to convert raw text into a clean list of tokens (words).

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
review_text.head()
```

Training the model

```python
model = gensim.models.Word2Vec(
    window = 10,        #context window size
    min_count = 2,      #minimum word frequency
    workers = 4         #number of CPU cores to use
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.epochs        # default number of iterations (epochs) over the corpus to train the model
```

```python
model.corpus_count   # number of sentences (reviews) in the corpus used to train the model
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# Testing the model - Finding similar words
model.wv.most_similar('good')
```

```python
model.wv.similarity(w1 = 'awesome', w2 = 'superb')
```

```python
model.wv.similarity(w1 = 'great', w2 = 'product')
```
